```
import tensorflow as tf
tf.__version__
```

```
mod_with_fc_raw = tf.keras.applications.efficientnet.EfficientNetB7(
    include_top=True, weights='imagenet', input_tensor=None,
    input_shape=(600,600,3), pooling='max', classes=1000
)
```

```
mod_with_fc_raw.summary()
```

```
for layer in mod_with_fc_raw.layers:
    layer.trainable = False 
```

```
mod_with_fc_raw.layers[-6].trainable = True 
```

```
mod_with_fc_raw.summary()
```

```
# Basically drop the dense layer which will be replaced with our residual muti - layer perceptron
# To avoid a potential conflict with the first BatchNormalization layer, I am also dropping the 
# dropout layer  
efficient_net_b_7_transferable_base_model =\
    tf.keras.Model(inputs=mod_with_fc_raw.layers[0].input, 
                   outputs=mod_with_fc_raw.layers[-3].output)
```

```
efficient_net_b_7_transferable_base_model.summary()
```

```
efficient_net_b_7_transferable_base_model.save('efficient_net_b_7_transferable_base_model')
```
